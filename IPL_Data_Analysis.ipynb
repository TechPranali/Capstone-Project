```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import csv

# Load the IPL dataset

df = pd.read_csv('IPL_Cricket_Project.csv', on_bad_lines='skip')
# Display the first few rows of the dataset
df.head()
```

```python
# Display basic info to understand the structure of the dataset
df.info()

# Check for missing values in the dataset
df.isnull().sum()
```

```python
# Count plot for toss decisions
plt.figure(figsize=(8, 5))
sns.countplot(x='toss_decision', data=df, palette='viridis')
plt.title('Distribution of Toss Decisions')
plt.xlabel('Toss Decision')
plt.ylabel('Count')
plt.show()
```

```python
# Count plot for match results by type (win by runs or win by wickets)
plt.figure(figsize=(10, 6))
sns.histplot(df['win_by_runs'], color='blue', label='Win by Runs', kde=True)
sns.histplot(df['win_by_wickets'], color='orange', label='Win by Wickets', kde=True)
plt.title('Distribution of Win Types')
plt.xlabel('Runs / Wickets')
plt.legend()
plt.show()
```

```python
# Bar plot for winning teams across different IPL seasons
plt.figure(figsize=(14, 7))
sns.countplot(x='season', hue='winner', data=df, palette='Spectral')
plt.title('Winning Teams by Season')
plt.xlabel('Season')
plt.ylabel('Win Count')
plt.legend(loc='upper right', bbox_to_anchor=(1.1, 1))
plt.xticks(rotation=45)
plt.show()
```
